```python
# Import our dependencies
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
from config import db_password
```

```python
# Read in Movies CSV
movies_df = pd.read_csv("Resources/movies.csv")
movies_df
```

```python
# Check for duplicates in the dataset movies_df
print(f"Duplicate entries: {movies_df['movie_name'].duplicated().sum()}")
```

```python
# Drop duplicates from movies_df
movies_df.drop_duplicates(subset="movie_name", keep=False, inplace=True)
movies_df
```

```python
# drop rows with Na values in movies_df
movies_df = movies_df.dropna()
movies_df
```

```python
#reset the index so that it is in order
movies_df = movies_df.reset_index(drop=True)
movies_df.count()
```

```python
# Read in Top Movies CSV
top_4000_movies_df = pd.read_csv("Resources/top_4000_movies_data.csv", parse_dates=['Release Date'], dayfirst=True)
top_4000_movies_df
```

```python
# Check for duplicates in the dataset top_4000_movies_df
print(f"Duplicate entries: {top_4000_movies_df['Movie Title'].duplicated().sum()}")
```

```python
# Drop duplicates from movies_df
top_4000_movies_df.drop_duplicates(subset="Movie Title", keep=False, inplace=True)
```

```python
# count number of rows
top_4000_movies_df.count()
```

```python
# Check for duplicates in the dataset top_4000_movies_df
print(f"Duplicate entries: {top_4000_movies_df['Movie Title'].duplicated().sum()}")
```

```python
# drop rows with Na values
top_4000_movies_df = top_4000_movies_df.dropna()
```

```python
# rename columns for 
top_4000_movies_df.rename(columns={'Release Date':'Release_Date', 'Movie Title':'Movie_Title', 'Production Budget':'Production_Budget', 'Domestic Gross':'Domestic_Gross',
       'Worldwide Gross':'Worldwide_Gross'},inplace=True)
```

```python
# reset the index so that it is in order
top_4000_movies_df.reset_index(drop=True)
```

```python
# count the number of rows
top_4000_movies_df.count()
```

```python
# create string to access postgreSQL database
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/movie_final"
```

```python
# create engine for sqalchemy
engine = create_engine(db_string)
```

```python
# export movies_df to movies table
movies_df.to_sql(name='movies', con=engine, index=False)
```

```python
# export top_4000_movies_df
top_4000_movies_df.to_sql(name='top_4000_movies', con=engine, index=False)
```
