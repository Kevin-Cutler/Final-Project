```python
# Import our dependencies
import pandas as pd
```

```python
# Read in Movies CSV
movies_df = pd.read_csv("Resources/movies.csv")
movies_df.head(20)
```

```python
# Check for duplicates in the dataset movies_df
print(f"Duplicate entries: {movies_df['movie_name'].duplicated().sum()}")
```

```python
# Drop duplicates from movies_df
movies_df.drop_duplicates(subset="movie_name", keep=False, inplace=True)
movies_df
```

```python
# drop rows with Na values in movies_df
movies_df = movies_df.dropna()
```

```python
movies_df.isnull()
```

```python
# Read in Top Movies CSV
top_4000_movies_df = pd.read_csv("Resources/top_4000_movies_data.csv", parse_dates=['Release Date'], dayfirst=True)
top_4000_movies_df.head(20)
```

```python
# Check for duplicates in the dataset top_4000_movies_df
print(f"Duplicate entries: {top_4000_movies_df['Movie Title'].duplicated().sum()}")
```

```python
# Drop duplicates from movies_df
top_4000_movies_df.drop_duplicates(subset="Movie Title", keep=False, inplace=True)
top_4000_movies_df
```

```python
# Check for duplicates in the dataset top_4000_movies_df
print(f"Duplicate entries: {top_4000_movies_df['Movie Title'].duplicated().sum()}")
```

```python
# drop rows with Na values
top_4000_movies_df.dropna()
```

```python
#export movies_df to a local csv file
movies_df.to_csv("Resources/movies_cleaned.csv", index=False)
```

```python
top_4000_movies_df.rename(columns={'Release Date':'Release_Date', 'Movie Title':'Movie_Title', 'Production Budget':'Production_Budget', 'Domestic Gross':'Domestic_Gross',
       'Worldwide Gross':'Worldwide_Gross'},inplace=True)
```
